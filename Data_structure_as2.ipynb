Q.1 What are data structures, and why are they important ?

Q.2 Explain the difference between mutable and immutable data types with examples ?
- Mutable data types can be modified after creation, while immutable data types cannot be changed once created. list, dictionary and set are mutales data types example while string and tuples are both immutable data types.

Q.5 Why might you use a set instead of a list in Python ?
- When uniqueness is required, I prefer using a set rather than a list.

Q.6 What is a string in Python, and how is it different from a list?
- A string is a data type in Python that stores characters and numbers as text, whereas a list can store multiple elements of different data types.


Q.7 How do tuples ensure data integrity in Python ?
- Tuples ensure data integrity in python as tuples are immutable. data modification is not possible in tuples.


Q.8 What is a hash table, and how does it relate to dictionaries in Python ?

Q.9 Can lists contain different data types in Python ?

Q.10 Explain why strings are immutable in Python ?

- Dictionaries offer faster lookup by key as dictionaries can not acceced by indexing so simply acceced by name(key) not by position.

Q.12 Describe a scenario where using a tuple would be preferable over a list ?
- A tuple is preferable over a list when storing fixed, read-only data where values should not change, ensuring data integrity and better performance.


Q.13 How do sets handle duplicate values in Python ?
- Sets store uniqe values so duplicatie values does no make any effect on set, and automatically remove this value.

Q.14 How does the “in” keyword work differently for lists and dictionaries?
- In lists, "in" checks for a value by element, while in dictionaries, "in" checks for a key.


Q.15 Can you modify the elements of a tuple? Explain why or why not?
- No, cannot modify the elements of a tuple because tuples are immutable in Python.

Q.17 Describe the time complexity of accessing elements in a dictionary ?
- Accessing elements in a dictionary is usually very fast because Python uses a hash table to find the value directly. In rare cases, when two keys try to occupy the same spot, finding the value can take longer.

Q.18 In what situations are lists preferred over dictionaries ?
- Lists are preferred over dictionaries when index-based access is required.

Q.19 Why are dictionaries considered unordered, and how does that affect data retrieval ?
- Dictionaries are considered unordered because they store items based on hash values, not positions, so data must be accessed via keys rather than relying on indexing.

Q.20 Explain the difference between a list and a dictionary in terms of data retrieval.
- Lists retrieve data by position (index) and may require lot of time, while dictionaries retrieve data by unique keys using hash tables, making access faster and more efficient.
